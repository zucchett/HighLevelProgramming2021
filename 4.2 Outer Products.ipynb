```python
import numpy as np
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
print(u)
print(v)
```

## Outer product using outer function

```python
print(np.outer(u,v))
```

## Outer product using nested for loop

```python
for i in u:
    row = []
    for j in v:
        row.append(i*j)
    print(row)
```
