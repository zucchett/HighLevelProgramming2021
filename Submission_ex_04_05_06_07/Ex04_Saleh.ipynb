1\. Find the total mean, and the mean for each row and column of the following matrix:

```python
m = np.arange(12).reshape((3,4))
```

````python
#Exercise 1 

m = np.arange(12).reshape((3,4))
print("Matrix: \n", m)
nrows, ncol = m.shape

#Total mean
mean = m.sum()/m.size
print("Total mean: ", mean)
  
#Mean of single rows
mean_row = m.sum(axis=1)/ncol
print("Mean in rows (0,1,2): ", mean_row)

#Mean of single columns
mean_col = m.sum(axis=0)/nrows
print("Mean in columns (0,1,2,3): ", mean_col)
````

2\. Find the outer product of the following vectors:

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

Use different methods to do this:

   1. Using the function `outer` in numpy
   2. Using a nested `for` loop or a list comprehension
   3. Using numpy broadcasting operations

````python
#Exercise 2

u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])

#2.1 - Outer function
matr= np.outer(u,v)
print("With outer function: \n", matr)

#2.2a - Nested for loop
n=len(u)
m=len(v)
matr = np.zeros((n,m))
for j in range(m):
    for i in range(n):
        matr[j][i] = u[j]*v[i]
print("\n With nested for loop: \n", matr)
        
#2.2b - Nested list comprehension
matr = [[u[j]*v[i] for i in range(n)] for j in range(m)]
print("\n With nested list comprehension: \n", matr)

#2.3 - Broadcasting operations
matr = u[:, None]*v[:]
print("\n With broadcasting operations: \n", matr)
````

5\. Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anti-diagonal matrix (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagonal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
#Exercise 5

#Create a matrix that shows the 10 by 10 multiplication table
a=np.arange(1,11)
b=np.arange(1,11)
m= a[:,np.newaxis]*b
print(m)

#Find the trace of the matrix
tr=np.trace(m)
print("\n Trace: ", tr)

#Extract the anti-diagonal matrix
anti = np.fliplr(m).diagonal()
print("\n Antidiagonal: ", anti)

#Extract the diagonal offset by 1 upwards
diag1= np.diagonal(m,1)
print("\n Diagonal with offset 1 upwards: ", diag1)
````

6\. Use broadcasting to create a grid of distances.

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles
The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Build a 2D grid of distances among each city along Route 66
  * Convert the distances in km

````python
#Exercise 6

import pandas as pd

distances=np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])
grid=distances-distances[:,np.newaxis]
print("Raw grid of distances (miles)")
print(grid)

print("\n \n Table of distances (miles) \n")
cities= np.array(["Chicago", "Springfield", "Saint-Louis", "Tulsa", "Oklahoma City", "Amarillo", "Santa Fe", "Albuquerque", "Flagstaff", "Los Angeles"])
grid = pd.DataFrame(grid, columns=cities, index=cities)
grid.columns.name = "City"
print(grid)

print("\n \n Table of distances (km) \n")
conv_factor = 1.609344 # 1 mile = 1.609344 km
grid = conv_factor*grid
print(grid)
````

7\. Prime numbers sieve: compute the prime numbers in the 0-N (start with N=99) range with a sieve (mask).
  * Check the performances (timeit); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
#Exercise 7

def prime(N):
    primes=[]
    for num in range(2,N):
        if all(num%i!=0 for i in range(2,num)):
            primes.append(num)
    return primes
    
#Compute the prime numbers in the 0-N (start with N=99) range with a sieve 
print("List of primes in [0,99]: \n",prime(100),"\n") 

#Check the performances (timeit); how does it scale with N?
print("Performances with n = 99")
%timeit prime(100)
print("\n")

print("Performances with n = 999")
%timeit prime(1000)
print("\n")

print("Performances with n = 9999")
%timeit prime(10000)
print("\n")


#Sieve of Eratosthenes
from math import sqrt

def eratosthenes(N):  
    is_prime = []
    for b in range(N):
        b = True
        is_prime.append(b)
    is_prime[0], is_prime[1] = False, False
    for i in range(2, int(sqrt(N))):
        for j in range(i**2, N, i):
            is_prime[j] = False
            
    primes=[]
    for i in range(2, N):
        if is_prime[i] == True:
            primes.append(i)
    
    return primes

print("Sieve of Eratosthenes: list of primes in [0,99] \n", eratosthenes(100), "\n")

print("Performances with n = 99")
%timeit eratosthenes(100)
print("\n")

print("Performances with n = 999")
%timeit eratosthenes(1000)
print("\n")

print("Performances with n = 9999")
%timeit eratosthenes(10000)
print("\n")
````
