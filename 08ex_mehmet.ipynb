1\. **2D minimization of a six-hump camelback function**

$$f(x,y) = \left(4-2.1x^2+\frac{x^4}{3} \right) x^2 +xy + (4y^2 -4)y^2$$

has multiple global and local minima.

- Find the global minima of this function
- How many global minima are there, and what is the function value at those points?
- What happens for an initial guess of $(x, y) = (0, 0)$?

Hints:

* Variables can be restricted to $-2 < x < 2$ and $-1 < y < 1$.
* Use `numpy.meshgrid()` and `pylab.imshow()` to find visually the regions.
* Use `scipy.optimize.minimize()`, optionally trying its optional arguments.

````python
#function definition with return function
def f(limitation):
    x, y = limitation
    return (4 - 2.1*x**2 + x**4/3.) * x**2 + x * y + (4*y**2 - 4) * y**2
#creating domain 
xes=np.linspace(-2, 2, 100)
yes=np.linspace(-1, 1, 100)
x,y=np.meshgrid(xes, yes)
# plot layout functions to display
fig = plt.figure(figsize=(20,10))
axis1 = plt.subplot2grid((1, 3), (0, 0), rowspan=1, colspan=2, projection='3d')
axis2 = plt.subplot2grid((1, 3), (0, 2), rowspan=1, colspan=1)
#3d plot function with maping coolwarm colors
axis1.plot_surface(x, y, f((x,y)), cstride=1, rstride=1, cmap='jet')
#heat map and contour for our plot graphics
im = axis2.imshow(f((x,y)), cmap='jet', origin='lower', extent=[-2, 2, -1, 1])
axis2.contour(x, y, f((x,y)), cmap='jet', levels=np.arange(-6, 6, 0.5))
#adding colobar
divider = make_axes_locatable(axis2)
cax2 = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im, cax=cax2, orientation='vertical')
#showing plot 
plt.show()
````

````python
'''after the minimum points are determined i set the limits and used the scipy.optimize.minimize() function'''
###specify bounds and  splits [-1,1] in half and [-2,2] in three parts###
bounds = [Bounds([-2, -1], [-1, 0]), Bounds([-1, -1], [1, 0]), Bounds([1, -1], [2, 0]), 
          Bounds([-2,  0], [-1, 1]), Bounds([-1,  0], [1, 1]), Bounds([1,  0], [2, 1])]
# our  initial starting point  is [x0, y0]
p0 = [[-1.5, -0.5], [0., -0.5], [1.5, -0.5],
      [-1.5,  0.5], [0.,  0.5], [1.5,  0.5]]
print("minimize the function below")
###unziping our bounds###
for guess, bound in zip(p0, bounds):
    #apply our algorithm for result
    res = sp.optimize.minimize(f, guess, bounds=bound)
    resPar = res.x
    x_min, y_min = resPar
    print("\nx min =", x_min,"\ty min =", y_min,
          "\tf(x min, y min) =", f(resPar))
````

````python
'''
    fin the minimum with (0,0) this is saddle point for our function for summary the minimization stucks in a local minima if we choose (0, 0) as starting point
'''

print("minimize func with starting points (0,0)")
p0 = [0., 0.]
res = sp.optimize.minimize(f, p0)
resPar = res.x
x_min, y_min = resPar
print("\nx min =", x_min,"\ty min =", y_min, 
      "\tf(x min, y min) =", f(resPar))
````

The temperature extremes in Alaska for each month, starting in January, are given by (in degrees Celcius):

* Plot these temperatures.
* Find a suitable a function that can describe min and max temperatures. 
* Fit this function to the data with `scipy.optimize.curve_fit()`.
* Plot the result. Is the fit reasonable? If not, why?
* Is the time offset for min and max temperatures the same within the fit accuracy?

````python
tempatureMaxx = np.array([17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18])
tempatureMinn = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
months = np.arange(1, 13)
##plotcodes tempature mesaurements##
fig, ax = plt.subplots(figsize=(16, 12))
ax.plot(months, tempatureMaxx, 'x', color='blue', label='max temp measurements')
ax.plot(months, tempatureMinn, 'x', label='min temp measurements')
ax.xaxis.set_major_locator(FixedLocator(range(13)))
ax.yaxis.set_major_locator(MultipleLocator(5))
##gauss function with return##
def gauss_function(x, a, mu, sigma, c):
    return a*np.exp(-(x-mu)**2/(2*sigma**2)) + c

popt_max, pcov = optimize.curve_fit(gauss_function, months, tempatureMaxx)
popt_min, pcov = optimize.curve_fit(gauss_function, months, tempatureMinn)
monthsHightsR = np.arange(1, 12.1, 0.1)
tempatureMaxx_fit = gauss_function(monthsHightsR, popt_max[0], popt_max[1], popt_max[2], popt_max[3])
tempatureMinn_fit = gauss_function(monthsHightsR, popt_min[0], popt_min[1], popt_min[2], popt_min[3])
##plot codes labels##
ax.plot(monthsHightsR, tempatureMaxx_fit, label='maximum tempature fit')
ax.plot(monthsHightsR, tempatureMinn_fit, color='yellow', label='mininimum tempatures fit')
ax.axvline(monthsHightsR[np.argmax(temp_max_fit)], color='darkblue', label='time offset maximum tempature')
ax.axvline(monthsHightsR[np.argmax(temp_min_fit)], color='pink', label='time offset minimum tempature')
ax.set_xlabel('Months')
ax.set_ylabel('Temperatures')
ax.legend()
````

````python
###The fit seems good for both the sets of measurements, capturing the seasonality of the trends. However, the time offset fot min and max temperatures differs###
print("Maximum temperature time offset -> {:.2f}".format(monthsHightsR[np.argmax(tempatureMaxx_fit)]))

print("Minimum temperature time offset -> {:.2f}".format(monthsHightsR[np.argmax(tempatureMinn_fit)]))
````

````python
import pickle
import pandas as pd
##opening file 
filename = "data/residuals_261.pkl"
infile = open(filename, 'rb') 
new_dict = pickle.load(infile) 
infile.close()

data=dict(new_dict.item())
df=pd.DataFrame(data)
#limited range to exclude outliers from the fit
df[abs(df.residuals)<3]
xl=-3
xu=3
fig, ax = plt.subplots(figsize=(10, 6)) # create the figure and the axes
yvalues, bins, _ = plt.hist(df['residuals'], bins=200, range=(xl, xu), cumulative=False) 
binwidth=bins[1]-bins[0]
bins=bins[:len(bins)-1]

ax.set_title('History',fontsize=20)
ax.set_xlabel('Residuals',fontsize=20)
ax.set_ylabel('Counts', fontsize=20)
ax.set_xlim(-3,3)
ax.errorbar(bins+binwidth/2, yvalues, yerr=np.sqrt(yvalues), marker='s', linestyle='None')
fig.tight_layout()
````

````python
#gausfunction definition with return##
def gaus(x, const, mu, sigma):
    return const * np.exp(-0.5*(x-mu)**2/sigma**2)

print("Doing Fit")
popt_r, pcov_r = optimize.curve_fit(gaus, bins+binwidth/2, yvalues, p0=[500,0,1], full_output=False)
print("\nOur parameters\n", popt_r)
print("\n\nUsing Covariance Matrix\n", pcov_r)
print("Const \t=\t %.2f \t+-\t %.2f" % (popt_r[0], np.sqrt(pcov_r[0,0])))
print("Mean value \t=\t %.3f \t+-\t %.3f" % (popt_r[1], np.sqrt(pcov_r[1,1])))
print("Sigma value \t=\t %.3f \t+-\t %.3f" % (popt_r[2], np.sqrt(pcov_r[2,2])))


# creating the figure and the axes

fig, ax = plt.subplots(figsize=(10, 6)) 

yvalues, bins, _ = plt.hist(df['residuals'], bins=200, range=(xl, xu), cumulative=False) 
binwidth=bins[1]-bins[0]
bins=bins[:len(bins)-1]
##axes
ax.set_title('History',fontsize=10)
ax.set_xlabel('Residuals',fontsize=10)
ax.set_ylabel('Counts', fontsize=10)
ax.set_xlim(-3,3)
ax.errorbar(bins+binwidth/2, yvalues, yerr=np.sqrt(yvalues), marker='s', linestyle='None')
ax.plot(np.arange(start=xl,stop=xu,step=0.1), gaus(np.arange(start=xl,stop=xu,step=0.1), popt_r[0], popt_r[1],popt_r[2]), label='fit',c='red')
ax.legend(loc='best')
fig.tight_layout()
````
