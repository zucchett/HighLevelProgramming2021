1\. PCA on 3D dataset

* Generate a dataset simulating 3 features, each with N entries (N being ${\cal O}(1000)$). Each feature is made by random numbers generated according the normal distribution $N(\mu,\sigma)$ with mean $\mu_i$ and standard deviation $\sigma_i$, with $i=1, 2, 3$. Generate the 3 variables $x_{i}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues using the eigendecomposition of the covariance matrix
* Find the eigenvectors and eigenvalues using the SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained
* Redefine the data according to the new basis from the PCA
* Plot the data, in both the original and the new basis. The figure should have 2 rows (the original and the new basis) and 3 columns (the $[x_0, x_1]$, $[x_0, x_2]$ and $[x_1, x_2]$ projections) of scatter plots.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.linalg as la
```

```python
N = 1000
np.random.seed(1)

a = np.random.normal(loc=0, scale=1, size=N)
b = np.random.normal(loc=0, scale=3, size=N) + a
c = 2 * a + b
X  = np.vstack((a, b, c))

cov = np.cov(X)

#eigenvalue decomposition of the covariance matrix
l, V = la.eig(cov)
l = np.real_if_close(l)
print('eigenvalues  cov(X):\n', l)
print('eigenvectors cov(X):\n', V)
print()

U, D, Vt = np.linalg.svd(X)
l_svd = D**2 / (N - 1)
V_svd = U
print('eigenvalues  from SVD:\n', l_svd)
print('eigenvectors from SVD:\n', V_svd)
print()

print('The results are the same')

# to select the best features
l = np.abs(l)
components = l / np.sum(l) * 100
for i in range(len(components)):
    print(f'variability principal component {i+1}: {components[i]:.2f}%')
print(f'if keeping the PC1 and PC3, retaining ~{components[0]+components[2]:.1f}% of the total variability')
print('second feature is simply the first feature affected by white noise')

# project the datapoints on the new basis of principal components
V_red = V[:,[0,2]]
X_red = np.matmul(V_red.T, X)

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.scatter(X_red[0,:], X_red[1,:], alpha=.7, s=25)
ax.set_title('Dimensionality reduction: from 3 to 2 features')
ax.set_ylabel('PC3')
ax.set_xlabel('PC1')

# change basis
Xp = np.matmul(V.T, X)

fig, axs = plt.subplots(2, 3, figsize=(10,6))
axs[0,0].scatter(X[0,:], X[1,:], alpha=.5, s=10, color='red', label='a vs b')
axs[0,1].scatter(X[0,:], X[2,:], alpha=.5, s=10, color='orange', label='a vs c')
axs[0,2].scatter(X[1,:], X[2,:], alpha=.5, s=10, color='black', label='b vs c')

axs[1,0].scatter(Xp[0,:], Xp[1,:], alpha=.5, s=10, color='red', label='PC1 vs PC2')
axs[1,1].scatter(Xp[0,:], Xp[2,:], alpha=.5, s=10, color='orange', label='PC1 vs PC3')
axs[1,2].scatter(Xp[1,:], Xp[2,:], alpha=.5, s=10, color='black', label='PC2 vs PC3')

for ax in axs.flatten():
    ax.legend(loc='upper left')
```

3\. **Optional**: PCA on the MAGIC dataset

Perform a PCA on the magic04.data dataset.

```python
columns = ["Length", "Width", "Size", "Conc", "Conc1", "Asym", "M3Long", "M3Trans", "Alpha", "Dist", "class"]

df = pd.read_csv('magic04.data', names=columns)
df = df.drop('class', axis=1)
df
```

```python
X = df - df.mean(axis=0, numeric_only=True) # center the variables around the mean
X = X.T # change to shape 
n = X.shape[1]
```

```python
# perform SVD
U, S, Vt = np.linalg.svd(X)

# rescale
l = S**2/(n-1)
V = U
```

```python
# to select the best features intrested only on the magnitude of the variability
components = l / np.sum(l) * 100
for i in range(len(components)):
    print(f'variability PC{i+1}: {components[i]:.2f}%')
print(f'{components[0]+components[1]+components[2]:.1f}% of the total variability')
```

```python
# project the datapoints 
V_red = V[:,[0,1,2]]
X_red = np.matmul(V_red.T, X)
```

```python
samples = 4000 # number of samples
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10,15))

ax1.scatter(X_red.iloc[0,:samples], X_red.iloc[1,:samples], alpha=.7, s=8, c='black')
ax1.axis([-250, 200, -150, 200])
ax1.set_ylabel('PC2')
ax1.set_xlabel('PC1')

ax2.scatter(X_red.iloc[0,:samples], X_red.iloc[2,:samples], alpha=.7, s=8, c='red')
ax2.axis([-250, 200, -200, 150])
ax2.set_ylabel('PC3')
ax2.set_xlabel('PC1')

ax3.scatter(X_red.iloc[1,:samples], X_red.iloc[2,:samples], alpha=.7, s=8, c='blue')
ax3.axis([-150, 200, -200, 150])
ax3.set_ylabel('PC3')
ax3.set_xlabel('PC2')
```
